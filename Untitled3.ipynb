```python
import json
import requests
import time
```

```python
class Name_finder:

    def __init__(self, document_url, max_seconds=60):
        root_url = 'http://gnrd.globalnames.org/name_finder.json'
        url = '{}?url={}&unique=true&data_source_ids=4'.format(root_url, document_url)
        res = requests.get(url).json()
        self.token_url = res['token_url']

        for s in range(1, max_seconds+1):
            time.sleep(1)
            print('\nPolling at {} seconds.\n'.format(s))
            self.json = requests.get(self.token_url).json()
            #print(self.json)
            if 'resolved_names' in self.json:
                break
```

```python
names_list = []
for item in self.json['resolved_names']:
    names_list.append(item['scientificName'])
names_set = set(names_list)
names_list = list(names_set)
names_list.sort()
self.names_list = names_list
```
