```python
import requests
import pandas as pd
import time
```

```python
class Name_finder:
    '''
    http://gnrd.globalnames.org/api
    '''

    def __init__(self, document_url, max_seconds=60):
        root_url = 'http://gnrd.globalnames.org/name_finder.json'
        url = '{}?url={}&unique=true&data_source_ids=11'.format(root_url, document_url)
        res = requests.get(url).json()
        self.token_url = res['token_url']

        for s in range(1, max_seconds+1):
            time.sleep(1)
            print('Polling at {} seconds.'.format(s))
            self.json = requests.get(self.token_url).json()
            #print(self.json)
            if 'resolved_names' in self.json:
                print('FINISHED - resolved names returned in self.json["resolved_names"]')
                print('token url = {}'.format(self.token_url))
                break
                
# document_url = 'https://aubreymoore.github.io/crop-pest-list/list.html'
# x = Name_finder(document_url)
```

```python
r = Name_finder('https://aubreymoore.github.io/crop-pest-list/list.html')
```

```python
taxa = r.json['resolved_names']
taxa[:2]
```

```python
mylist = []
for t in taxa:
    mydict = {'supplied_name_string': t['supplied_name_string'], 
              'is_known_name': t['is_known_name']}
    if 'results' in t:
#    if t['is_known_name']:
        mydict['taxon_id'] = t['results'][0]['taxon_id']
        mydict['match_value'] = t['results'][0]['match_value']
    mylist.append(mydict)
mylist 
df = pd.DataFrame(mylist).sort_values('supplied_name_string')
df
```

```python
pd.options.display.max_rows = 1000
df[df.is_known_name==False]
```

```python
# Delete rows where supplied_name_string is invalid
invalid_list = ['Abelmoschus esculentus chinese','Acari','Avocado','Brassica chinese','Phasmatodea','Rosa chinese',
                'Zea mays oriental','Zingiber officinale banana']
for name in invalid_list:
    df = df[df.supplied_name_string != name]
df[df.is_known_name==False]
```
